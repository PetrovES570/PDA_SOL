### Step 1. Import the necessary libraries

```python
import numpy as np
```

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
address = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(address)
```

```python
drinks.head(5)
```

### Step 4. Which continent drinks more beer on average?

```python
drinks[drinks['beer_servings'] > drinks['beer_servings'].mean()].groupby(['continent']).count()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby(['continent'])['wine_servings'].sum()
```

```python
drinks.groupby(['continent'])['wine_servings'].describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks[['continent', 'beer_servings', 'spirit_servings','wine_servings', 'total_litres_of_pure_alcohol']].groupby(['continent']).mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks[['continent', 'beer_servings', 'spirit_servings','wine_servings', 'total_litres_of_pure_alcohol']].groupby(['continent']).median()
```

```python
df_1 = drinks[['continent', 'beer_servings', 'spirit_servings','wine_servings', 'total_litres_of_pure_alcohol']].groupby(['continent']).median()
```

```python
df_1.to_csv (r'_.csv', index= True)
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
pd.DataFrame(data = {'mean' : drinks.groupby(['continent'])['spirit_servings'].mean(),
                    'min' : drinks.groupby(['continent'])['spirit_servings'].min(),
                    'max' : drinks.groupby(['continent'])['spirit_servings'].max()})
```
