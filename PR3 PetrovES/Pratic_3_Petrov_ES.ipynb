```python
import pandas as pd
import numpy as np
import json
import random
from faker import Faker

# Генерация данных о сотрудниках
fake = Faker()
qualifications = ['Junior', 'Middle', 'Senior']
employees = [
    {
        'name': fake.name(),
        'qualification': random.choice(qualifications),
        'experience': random.randint(1, 20)
    }
    for _ in range(100)
]

# Создание CSV-файла
employees_df = pd.DataFrame(employees)
employees_df.to_csv('employees.csv', index=False)

# Генерация данных о проектах
projects = {
    'project': [f'Project_{i}' for i in range(1, 11)] * 10, # 10 проектов, по 10 работ на каждый
    'employee': [random.choice(employees_df['name'].tolist()) for _ in range(100)],
    'hours_worked': [random.randint(1, 10) for _ in range(100)]
}

# Создание Excel-файла
projects_df = pd.DataFrame(projects)
projects_df.to_excel('projects.xlsx', index=False)

# Генерация данных о стоимости часа работы в зависимости от квалификации
hourly_rates = {
    'Junior': 20,
    'Middle': 35,
    'Senior': 50
}

with open('hourly_rates.json', 'w') as json_file:
    json.dump(hourly_rates, json_file)

print("Файлы созданы")
```

```python
# Чтение данных
employees_df = pd.read_csv('employees.csv')
projects_df = pd.read_excel('projects.xlsx')

# Загрузка hourly rates из JSON
with open('hourly_rates.json') as json_file:
    hourly_rates = json.load(json_file)

# Добавление стоимости часа для сотрудников в проекты
projects_df = projects_df.merge(employees_df, left_on='employee', right_on='name', how='left')
projects_df['hourly_rate'] = projects_df['qualification'].map(hourly_rates)

# Расчет стоимости работы по проекту
projects_df['total_cost'] = projects_df['hours_worked'] * projects_df['hourly_rate']

# Суммирование стоимости по каждому проекту
project_cost_summary = projects_df.groupby('project')['total_cost'].sum().reset_index()
project_cost_summary.columns = ['project', 'total_cost']

# Сохранение итогового результата в CSV
project_cost_summary.to_csv('project_cost_summary.csv', index=False)

print("Расчет общей стоимости работы по проектам в файле project_cost_summary")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка итогового результата
project_cost_summary = pd.read_csv('project_cost_summary.csv')

# Создание столбчатой диаграммы
plt.figure(figsize=(10, 6))
plt.bar(project_cost_summary['project'], project_cost_summary['total_cost'], color='skyblue')
plt.title('Общая стоимость работы по проектам')
plt.xlabel('Проект')
plt.ylabel('Общая стоимость')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Добавление стоимости по квалификациям в исходный DataFrame
qualification_costs = projects_df.groupby('qualification')['total_cost'].sum().reset_index()

# Создание круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(qualification_costs['total_cost'], labels=qualification_costs['qualification'], autopct='%1.1f%%', startangle=140)
plt.title('Распределение затрат по квалификациям сотрудников')
plt.axis('equal')  # Равные оси для равного круга
plt.show()
```
